```python
from sklearn import neighbors, svm
from torch import randperm
from torchvision.datasets import MNIST, FashionMNIST, CIFAR10
from sklearn.metrics import confusion_matrix, accuracy_score
from torch.utils.data.dataset import Dataset
import torch
from numpy.linalg import norm
import numpy as np
from scipy.sparse.csgraph import laplacian
from scipy.sparse.linalg import eigsh
from matplotlib import pyplot as plt

class PreprocessedDataset(Dataset):
    def __init__(self, dataset, max_size=None, to_tensor=False, split=None, seed=1):
        
        
#         print(dataset.train_data[0, :15, :15])
        
        self.seed = torch.manual_seed(seed)
        self.labeled = None
        self.unlabeled = None
        dataset = self.convert_data_attributes(dataset)
        
        if to_tensor:
            dataset = self.convert_numpy_to_tensor(dataset)
        
        if max_size is not None:
            dataset = self.reduce_size(dataset, max_size)
            
        if split is not None:
            dataset = self.split_unlabeled(dataset, split)
            
        self.data = dataset.data
        self.labels = dataset.labels
        
#         print(self.data[0, :15, :15])
        
    def convert_data_attributes(self, dataset):
        if dataset.train:
            dataset.data = dataset.train_data
            dataset.labels = dataset.train_labels
        else:
            dataset.data = dataset.test_data
            dataset.labels = dataset.test_labels
        return dataset
    
    def convert_numpy_to_tensor(self, dataset):
        dataset.data = torch.from_numpy(dataset.data)
        dataset.labels = torch.LongTensor(dataset.labels)
        return dataset
        
    def reduce_size(self, dataset, max_size):
        indices = randperm(len(dataset), generator=self.seed).tolist()
        dataset.data = dataset.data[indices[:max_size]]
        dataset.labels = dataset.labels[indices[:max_size]]
        return dataset
    
    def split_unlabeled(self, dataset, split):
        """Removes labels for split % of the dataset, replaces those labels with -1"""
        x = dataset.data
        y = dataset.labels
        # Split train into labeled/unlabeled, simulating semi-supervised learning
        self.labeled = int(split * len(x))
        self.unlabeled = len(x) - self.labeled
        lengths = [self.labeled, self.unlabeled]
        # Randomize the indices
        indices = randperm(sum(lengths), generator=self.seed).tolist()
#         print(x.type())
        x_labeled = x[indices[:self.labeled]]
#         print(y.type())
#             print(y.shape)
#             print(indices.shape)
#             print(indices[:self.labeled].shape)
        y_labeled = y[indices[:self.labeled]]
        x_unlabeled = x[indices[-self.unlabeled:]]
        y_unlabeled = torch.ones(self.unlabeled).long() * -1 # -1 indicates to UMAP a missing label - See https://umap-learn.readthedocs.io/en/latest/supervised.html#using-partial-labelling-semi-supervised-umap
    
        dataset.data = torch.cat((x_labeled, x_unlabeled))
        dataset.labels = torch.cat((y_labeled, y_unlabeled))
        
        return dataset
    
    def __getitem__(self, i):
        return self.data[i], self.labels[i]
    
    def __len__(self):
        return len(self.labels)
    
class Graph():
    """Creates the graph specified (UMAP, KNN, etc.) Obtains the Laplacian
    of that graph and the eigenvectors of the Laplacian. The eigenvectors are
    basis functions on the graph and the entry eig_vecs[pt, vec] gives the
    coordinate of 'pt' in terms of the basis function 'vec'. Thus 
    eig_vecs[pt, :] gives the representation of a point in terms of basis
    functions on the graph. This object provides this representation for seen
    points and unseen points."""
    def __init__(self, dataset, n_neighbors, graph_type='knn', metric='minkowski', n_eig_vecs=10):
        
#         print(dataset.data[0, :15, :15])
        self.labeled = dataset.labeled
        self.unlabeled = dataset.unlabeled
        # Most graph creation methods only operate on vectors, not matrices (like images)
        if len(dataset.data.size()) == 4:
#             data_reshaped = dataset.data.contiguous()
#             print(data_reshaped.size())
            data_reshaped = dataset.data.contiguous().view(dataset.data.shape[0], -1)
#             print(data_reshaped.size())
        else:
            data_reshaped = dataset.data.view(dataset.data.shape[0], -1)
        
#         print(data_reshaped[0, :15])
        
        # TODO: Find other comparisons for graph creation. A fundamental 
        # weakness of the KNN graph is that it doesn't account for categorical
        # features (which gives UMAP a huge leg up)
        # Make the graph
        if graph_type == 'knn':
            adj_matrix = neighbors.kneighbors_graph(
                data_reshaped, 
                n_neighbors, 
                mode='connectivity',
                metric=metric, 
                p=2, 
                metric_params=None, 
                include_self=False, 
                n_jobs=None
            )
#             plt.imshow(adj_matrix[:100, :100].todense())
        else:
            assert graph_type == 'umap'
            # TODO: Change the next line to UMAPGraph(), so that the optimization process is skipped
#             umap = UMAP(n_components=2, n_neighbors=n_neighbors)
#             umap.fit_transform(data.data, data.labels)
#             adj_matrix = umap.graph_
            adj_matrix = points_to_graph(dataset.data, dataset.labels)

        # Symmetrize the adjacency matrix
        self.adj_matrix = np.maximum( adj_matrix.todense(), adj_matrix.todense().T )

        laplace = laplacian(self.adj_matrix, normed=True)

        # eigsh gets eigenvectors and values from the graph laplacian
        _, eig_vecs = eigsh(laplace, k=n_eig_vecs+1, which='SM', maxiter=5000)
        
        # TODO: What should these be called? They aren't exactly embeddings
        self.embeddings = eig_vecs[:, 1:]
        
        # We only have labels for x_labeled
        self.labels = dataset.labels
        
        # TODO: Provide different metrics for neighbors (VGG cosine)
        # This is used in the embed_new_pts() method for finding the graph coordinates of new points
        self.ngbrs = neighbors.NearestNeighbors(n_neighbors=3, metric=metric, p=2).fit(data_reshaped)
        
    def get_pts(self, which='labeled'):
        """Gets the embeddings of the original points in the graph. Careful - 
        even unlabled points have a 'label' of -1, indicating that they are 
        unlabeled"""
        if which == 'labeled':
            return self.embeddings[:self.labeled], self.labels[:self.labeled]
        elif which == 'unlabeled':
            return self.embeddings[self.labeled:], self.labels[self.labeled:]
        else:
            assert which == 'all'
            return self.embeddings, self.labels
    
    def embed_new_pts(self, x):
        """Finds the graph coordinates of an unseen point"""
#         print(type(x))
        data_reshaped = x.view(x.shape[0], -1)
        distances, indices = self.ngbrs.kneighbors(data_reshaped)
        # Find the graph coordinates of the new point's nearest neighbors
        nearest_neighbors = self.embeddings[indices, :]
        # Let the new point be the average of its neighbors
        return np.mean(nearest_neighbors, axis=1)
        

class Classifier():
    def __init__(self, 
                 graph,
                 classifier_type='knn', # ['knn', 'svc']
                 n_neighbors=5          # Only necessary if classifier_type == 'knn'
                ):

        self.graph = graph
        # Transform the original points into their graph embedding
        x_labeled, y_labeled = graph.get_pts(which='labeled')

        if classifier_type == 'knn':
            # Make knn classifier
            self.classifier = neighbors.KNeighborsClassifier(n_neighbors=n_neighbors)
            print("KNN: ", x_labeled.shape)
            self.classifier.fit(x_labeled, y_labeled)
        else:
            assert classifier_type == 'svc'
            # Make suport vector classifier
            self.classifier = svm.SVC()
            print("SVC: ", x_labeled.shape)
            self.classifier.fit(x_labeled, y_labeled)
        
    def predict(self, x):
        # Predict
        x_embedded = self.graph.embed_new_pts(x)
        y_pred = self.classifier.predict(x_embedded)
        return y_pred
        
def plot_acc(y_truth, y_pred, dataset, graph, classifier):
        # Get accuracy
        acc = accuracy_score(y_truth, y_pred)
        # Make confusion matrix
        conf = confusion_matrix(y_truth, y_pred)
        # Report Accuracy, type of graph
        plt.imshow(conf)
        plt.title(f"Confusion Matrix for {dataset} dataset based on {graph} graph with \n{classifier} classifier. Accuracy: {100*acc}%")
```
